```python
# import library
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots that will create
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# check the missing value
for column in df.columns :
    percent_missing_value = np.mean(df[column].isnull())
    print('{} - {}%'.format(column, percent_missing_value))
```

```python
df.isnull().sum()
```

```python
df1 = df.dropna()
```

```python
df1.isnull().sum()
```

```python
# check data types
df1.dtypes
```

```python
# changing datatype
df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
```

```python
df1.dtypes
```

```python
df1.head()
```

```python
#sorting the data
df1 = df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop any duplicates
df1['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df1.head()
```

```python
# hypothesis 1: 'budget' has high correlation with 'gross'
# scatter plot --> budget vs gross

plt.scatter(x=df1['budget'], y=df1['gross'])

plt.title('Budget vs Gross')
plt.xlabel('Gross')
plt.ylabel('Budget')

plt.show()
```

```python
# plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df1, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
df1.corr().head()
```

```python
# from the result above, it confirm that 'budget' has high correlation with 'gross' = 0,74
```

```python
correlation_matrix = df1.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```

```python
# hypothesis 2: 'company' has high correlation with 'gross'
# because 'company' is not numeric, we wil numerize the datafame 
# so we can make correlation with company and gross

df_numerized = df1
for col_name in df_numerized.columns :
    if (df_numerized[col_name].dtype == 'object'):
         df_numerized[col_name] = df_numerized[col_name].astype('category')
         df_numerized[col_name] = df_numerized[col_name].cat.codes
            
df_numerized.head()
```

```python
df_numerized.corr().head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```

```python
# from the result above, it not that 'budget' has high correlation with 'gross' 
# because the correlation is low = 0,15
```
